#### 06.결측치 제거

```python
import pandas

df = pandas.read_excel('test1.xlsx')
df
```

```python
# 결측치 : 데이터 세트에서 누락된 값 (ex> 비어있는 데이터, NaN 데이터)
```

```python
# 결측치를 제거하는 방법은 결측치를 삭제하는 방법, 결측치를 특정 데이터로 채우는 방법
```

```python
# 실습을 위해 결측치 데이터를 추가
import numpy

df['학급'] = numpy.nan
df
```

```python
# NaN 데이터를 특정 데이터로 채우기
df.fillna('임시데이터', inplace=True)
```

```python
df
```

```python
# 특정 컬럼의 NaN을 수정하기
df['취미'].fillna('없음', inplace=True)
```

```python
# NaN 데이터 제거
# dropna의 매개변수
# axis = 'index' : NaN 데이터를 기준으로 행을 삭제(default)
# axis = 'columns' : NaN 데이터를 기준으로 열을 삭제 
# how = 'any' : NaN 데이터가 하나라도 존재하면 삭제(default)
# how = 'all' : 모든 데이터가 NaN일 때 삭제
df.dropna(axis='index' , how='any', inplace=True)
```
