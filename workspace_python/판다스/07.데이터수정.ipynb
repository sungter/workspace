#### 07. 데이터 수정

```python
import pandas

df = pandas.read_excel('test1.xlsx')
df
```

```python
# 모든 학생의 영어점수를 현재 점수 + 5
df['영어점수'] = df['영어점수'] + 5

df
```

```python
# 평균 점수 수정
df['평균'] = (df['국어점수'] + df['영어점수']) // 2

df
```

```python
# 컬럼 추가하기
df['결과'] = 'Fail'
df
```

```python
# 특정 조건에 맞는 데이터만 수정
# 국어 , 영어 합이 170점 이상
df.loc[df['국어점수'] + df['영어점수'] >= 170 , '결과'] = 'Pass'
df
```

```python
# 컬럼 삭제하기
# df.drop(columns={'결과'}) # 결과 컬럼 삭제
df.drop(columns={'평균', '결과'}) # 평균, 결과 컬럼 삭제 // inplace=True까지 확정
```
